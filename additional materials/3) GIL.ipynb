# GIL - это глобальный блокировщик интерпретатора, он позволяет работать лишь одному потоку в конкретный момент времени

### Одно ядро процессора может исполнять только один поток
### Чтобы обойти гил, используют несколько процессов через multiprocessing, дописывают расширения на С или используют интерпретаторы Juthon или IronPython
### Многопоточность в Python увеличивает скорость работы только в IO-bound задачах (запрост к БД, загрузка инфы, вывод в терминал)
### Процесс - исполняемый файл (программа) и ресурс для неё (память и процессор)
### Поток - единица выполнения внутри процесса. Один процесс может иметь несколько потоков. Потоки используют общую память процесса, но имеют свой стек
### Один процесс не может получить доступ к другому. Для обмена данными между ними нужен Celery или multiprocessing
